```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
credit_data = pd.read_csv('tmdb_5000_credits.csv')
movie_data = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credit_data.head()
```

```python
movie_data.head()
```

```python
credit_data.shape
```

```python
movie_data.shape
```

```python
credit_data = credit_data.rename(index=str,columns={"movie_id":"id"})
movie_merge = movie_data.merge(credit_data,on='id')
movie_merge.head()
```

```python
movie_merge.shape
```

```python
movie_merge.columns
```

```python
movie_merge.info()
```

```python
movie_merge = movie_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movie_merge.columns
```

```python
v = movie_merge['vote_count']
R = movie_merge['vote_average']
C = movie_merge['vote_average'].mean()
m = movie_merge['vote_count'].quantile(0.7)
```

```python
movie_merge['weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
movie_merge['weighted_average']
```

```python
sorted_movie = movie_merge.sort_values('weighted_average',ascending=False)
sorted_movie[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
plt.figure(figsize=(12,6))
ax1 = sns.barplot(x=sorted_movie['weighted_average'].head(10),y=sorted_movie['original_title'].head(10),data=sorted_movie)
plt.xlim(4,10)
plt.title('Best movie by average votes',weight='bold')
plt.xlabel('Weighted average score',weight='bold')
plt.ylabel('Movie title',weight='bold')
```

```python
popularity = sorted_movie.sort_values('popularity',ascending=False)
```

```python
popularity.head()
```

```python
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
movie_scaled = scale.fit_transform(movie_merge[['weighted_average','popularity']])
movie_normalized = pd.DataFrame(movie_scaled,columns=['weighted_average','popularity'])
movie_normalized.head()
```

```python
movie_merge[['normalized_weighted_average','normalized_popularity']] = movie_normalized
```

```python
movie_merge.head()
```

```python
movie_merge['score'] = movie_merge['normalized_weighted_average']*0.5 + movie_merge['normalized_popularity']*0.5
```

```python
movie_merge = movie_merge.sort_values(['score'],ascending=False)
```

```python
movie_merge[['original_title','normalized_weighted_average','normalized_popularity','score']].head(15)
```

```python
score = movie_merge.sort_values('score',ascending=False)
plt.figure(figsize=(12,6))

ax = sns.barplot(x=score['score'].head(10),y=score['original_title'].head(10),data=score,palette='deep')

plt.title('Best Rated & Most Popular',weight='bold')
plt.xlabel('Score')
plt.ylabel('Movie Title')
```
