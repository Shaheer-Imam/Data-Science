```python
data.drop(data[data['num_lectures']==0].index,inplace=True)
```

```python
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])
data['published_date'] = data['published_timestamp'].dt.date
data['published_year'] = pd.DatetimeIndex(data['published_timestamp']).year
```

### Top 10 Most Subscribed Courses in terms of Subjects

```python
subjects = ['Business Finance', 'Graphic Design', 'Musical Instruments',
           'Web Development']
for i in subjects:
    plt.figure(figsize=(10,4))
    sns.barplot(data=data[data['subject']==i].nlargest(10,'num_subscribers'),x='num_subscribers',y='course_title')
    plt.title("Top 10 Most Subscribed " +i+ " Courses")
    plt.show()
```
